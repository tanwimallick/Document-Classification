# Stronger embedding models validation
In this experiment, we use the top models from the Huggingface's MTEB. Some of these are quite large, so we run them on Argonne's Polaris supercomputer and download the results to classify using Snorkel and validate here.

```python
from snorkel.labeling import labeling_function
from snorkel.labeling.model import LabelModel
from snorkel.labeling import PandasLFApplier
import pandas as pd
from sentence_transformers import SentenceTransformer, util
```

**Key**:
- LF_1: Salesforce/SFR-Embedding-Mistral
- LF_2: Labib11/MUG-B-1.6
- LF_3: Linq-AI-Research/Linq-Embed-Mistral (formerly LF_4)
- LF_4: intfloat/multilingual-e5-large-instruct (formerly LF_5)
- LF_5: mixedbread-ai/mxbai-embed-large-v1 (formerly LF_6)

```python
d1 = pd.read_csv("data/lf/lf_1_GPTLong.csv")
d2 = pd.read_csv("data/lf/lf_2_GPTLong.csv")
d3 = pd.read_csv("data/lf/lf_3_GPTLong.csv")
d4 = pd.read_csv("data/lf/lf_4_GPTLong.csv")
d5 = pd.read_csv("data/lf/lf_5_GPTLong.csv")
```

```python
df = d1.merge(d2, on=['ID', 'TITLE', 'ABSTRACT', 'Computer Science', 'Physics', 'Mathematics', 'Statistics', 'Quantitative Biology', 'Quantitative Finance'], how='outer') \
            .merge(d3, on=['ID', 'TITLE', 'ABSTRACT', 'Computer Science', 'Physics', 'Mathematics', 'Statistics', 'Quantitative Biology', 'Quantitative Finance'], how='outer') \
            .merge(d4, on=['ID', 'TITLE', 'ABSTRACT', 'Computer Science', 'Physics', 'Mathematics', 'Statistics', 'Quantitative Biology', 'Quantitative Finance'], how='outer') \
            .merge(d5, on=['ID', 'TITLE', 'ABSTRACT', 'Computer Science', 'Physics', 'Mathematics', 'Statistics', 'Quantitative Biology', 'Quantitative Finance'], how='outer')
```

**Compute Snorkel labels**

Using the above graphs as a loose guide, we devise labeling functions that we combine in a Snorkel weak supervision model

```python
@labeling_function()
def lf_def_1_CS(x):
  score = df[df['ABSTRACT'] == x['ABSTRACT']]['LF_1_CS'].values[0]
  if score > 0.5:
   return 1  
  return 0

@labeling_function()
def lf_def_2_CS(x):
  score = df[df['ABSTRACT'] == x['ABSTRACT']]['LF_2_CS'].values[0]
  if score > 0.5:
   return 1  
  return 0

@labeling_function()
def lf_def_3_CS(x):
  score = df[df['ABSTRACT'] == x['ABSTRACT']]['LF_3_CS'].values[0]
  if score > 0.4:
   return 1  
  return 0

@labeling_function()
def lf_def_4_CS(x):
  score = df[df['ABSTRACT'] == x['ABSTRACT']]['LF_4_CS'].values[0]
  if score > 0.75:
   return 1  
  return 0

@labeling_function()
def lf_def_5_CS(x):
  score = df[df['ABSTRACT'] == x['ABSTRACT']]['LF_5_CS'].values[0]
  if score > 0.5:
   return 1  
  return 0

@labeling_function()
def lf_def_6_CS(x):
  score = df[df['ABSTRACT'] == x['ABSTRACT']]['LF_6_CS'].values[0]
  if score > 0.6: # Major decrease with 0.6, 0.65 might be better
   return 1  
  return 0
```

```python
lfs = [lf_def_1_CS, lf_def_2_CS, lf_def_3_CS, lf_def_4_CS, lf_def_5_CS, lf_def_6_CS]
```

```python
# Apply the LFs to the unlabeled training data
applier = PandasLFApplier(lfs)
L_train = applier.apply(df)

# Initialize the LabelModel
label_model = LabelModel(cardinality=2, verbose=True)

# Train the LabelModel
label_model.fit(L_train, n_epochs=500, log_freq=100, seed=123)
```

```python
# Get the predicted labels (y_train) for the data points
y_train = label_model.predict(L=L_train)
y_train
```

```python
df['Snorkel_CS'] = y_train
```

**Validate Snorkel**

```python
# Source: https://medium.com/@dtuk81/confusion-matrix-visualization-fc31e3f30fea

from sklearn.metrics import confusion_matrix
import seaborn as sns
import numpy as np

cm = confusion_matrix(df['Computer Science'], df['Snorkel_CS'])

group_names = ['True Neg','False Pos','False Neg','True Pos']
group_counts = ["{0:0.0f}".format(value) for value in
                cm.flatten()]
group_percentages = ["{0:.2%}".format(value) for value in
                     cm.flatten()/np.sum(cm)]
labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in
          zip(group_names,group_counts,group_percentages)]
labels = np.asarray(labels).reshape(2,2)

sns.heatmap(cm, annot=labels, fmt="", cmap='Blues')
```

```python
tn = cm[0][0]
fp = cm[0][1]
fn = cm[1][0]
tp = cm[1][1]

# Accuracy: the percent of correctly classified data
accuracy = (tn+tp)/(tn+fp+fn+tp)
accuracy
```

```python
# Precision: positive predictive value; of those predicted as CS documents, how many were actually CS documents?
precision = (tp)/(tp+fp)
precision
```

```python
# Recall: of those that were actually CS documents, how many were predicted as CS documents?
recall = (tp)/(tp+fn)
recall
```

```python
# F1 score: tries to balance precision and recall, with uneven class distribution (large number of actual negs)
f1 = 2*((precision*recall)/(precision+recall))
f1
```

```python
# Source: https://medium.com/@dtuk81/confusion-matrix-visualization-fc31e3f30fea

from sklearn.metrics import confusion_matrix
import seaborn as sns
import numpy as np

cm = confusion_matrix(df['Quantitative Finance'], df['Snorkel_QF'])

group_names = ['True Neg','False Pos','False Neg','True Pos']
group_counts = ["{0:0.0f}".format(value) for value in
                cm.flatten()]
group_percentages = ["{0:.2%}".format(value) for value in
                     cm.flatten()/np.sum(cm)]
labels = [f"{v1}\n{v2}\n{v3}" for v1, v2, v3 in
          zip(group_names,group_counts,group_percentages)]
labels = np.asarray(labels).reshape(2,2)

sns.heatmap(cm, annot=labels, fmt="", cmap='Blues')
```

```python
tn = cm[0][0]
fp = cm[0][1]
fn = cm[1][0]
tp = cm[1][1]

# Accuracy: the percent of correctly classified data
accuracy = (tn+tp)/(tn+fp+fn+tp)
accuracy
```

```python
# Precision: positive predictive value; of those predicted as CS documents, how many were actually CS documents?
precision = (tp)/(tp+fp)
precision
```

```python
# Recall: of those that were actually CS documents, how many were predicted as CS documents?
recall = (tp)/(tp+fn)
recall
```

```python
# F1 score: tries to balance precision and recall, with uneven class distribution (large number of actual negs)
f1 = 2*((precision*recall)/(precision+recall))
f1
```
